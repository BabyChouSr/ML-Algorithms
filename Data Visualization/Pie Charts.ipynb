```python
import numpy as np  
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
slices  = [10,20,30,40]
labels = ['TEN','TWENTY','THIRTY','FORTY']
colors = ['blue','green','yellow','purple']

plt.style.use('dark_background')
plt.pie(slices, labels = labels, colors = colors)
plt.title("pies are good")
plt.tight_layout()
```

```python
explode = [0,0,0.2,0]
slices  = [10,20,30,40]
labels = ['TEN','TWENTY','THIRTY','FORTY']
colors = ['blue','green','yellow','purple']

plt.style.use('dark_background')
plt.pie(slices, labels = labels, explode = explode, colors = colors, autopct = '%1.2f%%')
plt.title("pies are good")
plt.tight_layout()
```
