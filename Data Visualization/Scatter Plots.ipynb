```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x =  np.random.rand(10) * 10
y = np.random.rand(10) * 10
plt.style.use('bmh')
plt.scatter(x,y,s = 300, c= '#32a879',label = "Randomness")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Scattered")
plt.legend()
plt.show()
```

```python
x =  np.random.rand(10) * 10
y = np.random.rand(10) * 10

color = np.random.rand(10) * 10
plt.style.use('bmh')
plt.scatter(x,y,s = 300, c= color,cmap = 'Blues',label = "Randomness")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Scattered")
cbar = plt.colorbar()
cbar.set_label('Satisfaction Rating')
plt.legend()
plt.show()
```

```python
x =  np.random.rand(100) * 100
y = np.random.rand(100) * 1000
color = np.random.rand(100) * 10
size = np.random.rand(100) * 300 + 50
plt.style.use('bmh')
plt.scatter(x,y,s = size, c= color,cmap = "viridis",label = "Randomness")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Scattered")
plt.legend()
plt.show()
```
