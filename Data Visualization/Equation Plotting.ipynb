```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = [-4,-3,-2,-1,0,1,2,3,4]
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-4,4)
y = np.cos(x)
y2 = np.sin(x)
plt.style.use("ggplot")
plt.plot(x,y, label = "cos(x)")
plt.plot(x,y2, label = "sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of cos(x) and sin(x)")
plt.legend()
plt.show()
```
