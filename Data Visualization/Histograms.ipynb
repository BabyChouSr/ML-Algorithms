```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('ggplot')
x = np.random.rand(1000) * 10
bins = np.linspace(0,10,num = 100)
plt.hist(x, bins = bins, edgecolor = 'black',color = "#a8329e", label = "hey dummy")
plt.title("Dummy Data isn't Smart")
plt.xlabel("Ages of Dummy")
plt.ylabel("Frequency of Dummy")
plt.legend()
```

```python
xnorm = np.random.normal(0, 1, 1000)
bin_norm = np.linspace(-3,3,num=100)
plt.hist(xnorm, bins = bin_norm, edgecolor = 'black',color = "#42c8f5", label = "hey dummy2")
plt.title("Dummy Data isn't Smart")
plt.xlabel("Ages of Dummy")
plt.ylabel("Frequency of Dummy")
plt.legend()
```

```python
xnorm = np.random.normal(0, 1, 1000)
bin_norm = np.linspace(-3,3,num=100)
plt.hist(xnorm, bins = bin_norm, edgecolor = 'black',color = "#45de73", label = 'hey dummy3')
median = 0
plt.axvline(median, color = 'red', label = "Median")
plt.title("Dummy Data isn't Smart")
plt.xlabel("Ages of Dummy")
plt.ylabel("Frequency of Dummy")
plt.legend()
```
