```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = datasets.load_iris()
X = data['data']
y = data['target'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state = 20)
ans = []
xnum = []
for  i in range(1,50):
    neigh = KNeighborsClassifier(n_neighbors = i)
    neigh.fit(X_train,y_train)
    pred = neigh.predict(X_test)
    ans.append(accuracy_score(y_test,pred))
    xnum.append(i)

print("The Highest Accuracy is With ",(ans.index(max(ans))+1), " Neighbors with ", max(ans))
plt.plot(xnum,ans, label = "KNearestNeighbors")
```

```python
neigh2 = KNeighborsClassifier(n_neighbors = 7)
neigh2.fit(X_train,y_train)
pred2 = neigh2.predict(X_test)
print(pred2)
```

```python
plt.plot(X_test[:,0][pred2 == 0],X_test[:,1][pred2 == 0], "g.", label = "Iris-Setosa")
plt.plot(X_test[:,0][pred2 == 1],X_test[:,1][pred2 == 1], "r.", label = "Iris-Versicolour")
plt.plot(X_test[:,0][pred2 == 2],X_test[:,1][pred2 == 2], "b.", label = "Iris-Virginica")
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```python
plt.plot(X_test[:,0][y_test == 0],X_test[:,1][y_test == 0], "g.", label = "Iris-Setosa")
plt.plot(X_test[:,0][y_test == 1],X_test[:,1][y_test == 1], "r.", label = "Iris-Versicolour")
plt.plot(X_test[:,0][y_test == 2],X_test[:,1][y_test == 2], "b.", label = "Iris-Virginica")
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```
